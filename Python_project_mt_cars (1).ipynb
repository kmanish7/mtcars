```python
cd = cd.select_dtypes(include=['float64'])
print(cd)
```

```python
def identify_outliers(x):
    q1, q3 = np.percentile(x, [25, 75])
    iqr = q3 - q1
    lower_limit = q1 - (1 * iqr)
    upper_limit = q3 + (1 * iqr) 
    l = []
    for r in x:
        if (r < lower_limit):
             l.append('background-color: #FF3030')
        elif (r > upper_limit):
             l.append('background-color: #00C957')
        else:
            l.append('background-color: #0000CD')
    return l
cd.style.apply(identify_outliers)
```
